```python
import json
import numpy as np
import pandas as pd
from faker import Faker
from datetime import datetime, timedelta

fake = Faker()

# Number of fake users to generate
num_users = 100

# Generate fake data
data = []
start_date = datetime(2020, 1, 1)  # Starting point for fake dates

for i in range(num_users):
    # Generate a random number of days to add (between 0 and 1000)
    random_days = np.random.randint(0, 1000)
    fake_date = start_date + timedelta(days=int(random_days))
    
    user = {
        "user_id": i + 1,  # Random user ID between 1000 and 9999
        "name": fake.name(),
        "email": fake.email(),
        "date": fake_date.strftime("%Y-%m-%d")  # Format date as string
    }
    data.append(user)

# Write to JSON file
#with open("users.json", "w") as f:
#json.dump(data, f, indent=4)

df_user = pd.DataFrame(data)

# Display the DataFrame
df_user.head()
# 

```

```python
import json
import pandas as pd 
import numpy as np
from faker import Faker
from datetime import datetime, timedelta
import faker_commerce 

fake = Faker()

fake.add_provider(faker_commerce.Provider)

fake.ecommerce_name()

num_products = 100




# Generate fake data
data = []

for i in range(num_products):

    product = {
        "product_id": i + 1,  # Sequential product ID starting from 1
        "product_name": fake.ecommerce_name(),
        "category": np.random.choice(['Computers', 'Smartphone', 'Hardware', 'Monitor', 'Tools', 'Kitchen', 'Furniture']),
        "price": round(np.random.uniform(100, 500))  # Random price between 100 and 500
    }
    data.append(product)

# Write to JSON file
#with open("product.json", "w") as f:
#json.dump(data, f, indent=4)
 
# Convert the list of product dictionaries to a pandas DataFrame
df_products = pd.DataFrame(data)

# Display the DataFrame
df_products.head()
```

```python
df = pd.merge(df_user, df_products, left_on='user_id', right_on='product_id', how='inner')

df 
```
